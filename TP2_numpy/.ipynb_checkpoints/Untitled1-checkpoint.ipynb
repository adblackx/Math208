# Matplotlib
* sert à tracer des fonction et afficher différents graphique
* courbe d'une fonction réparamétre, camembers, histogrammes,représentation de signal 2D ou 3D
* on utilise la nouvelle interface: **pas de préfixage par plt** sauf pour crée la figure et les axes et le show

## Premier exemple minimalise

```python
#Chargement des modules 
import numpy as np
import matplotlib.pyplot as plt

#definition de ce qu'il faut tracer
x = np.linspace(-1,1)
y=x**2

#Graphe
fig, ax = plt.subplots(1,1) # creatiion figure et axe
ax.plot(x,y)
plt.show()
```

### Vers des exemplesplus complets
#### Tracer plusieurs focntion sur un meme graphe

```python
fig, ax = plt.subplots(1,1) # creatiion figure et axe
ax.plot(x,y)
ax.plot(x,y+0.5)
ax.plot(x,np.cos(x))

plt.show()
```

 test

```python
fig, ax = plt.subplots(1,1) # creatiion figure et axe
ax.plot(x,y,'g+', label="x^2")
ax.plot(x,np.cos(x), 'r*', label="cosinus")
ax.legend()
ax.set_title("x^2 et cosinus") # ajoute un titre
ax.set_xlabel("abscisses")
ax.set_ylabel("abscisses")

plt.show()
```

### Version plus evoluee

```python
fig, [ax, ax1] = plt.subplots(2,1, figsize=(10,10)) 

ax.plot(x,y,'g+', label="x^2")
ax.plot(x,np.cos(x), 'r*', label="cosinus")
ax.legend()
ax.set_title("x^2 et cosinus") # ajoute un titre
ax.set_xlabel("abscisses")
ax.set_ylabel("ordonnées")

ax1.plot(x,np.tan(x),'g+', label="tangente")
ax1.set_title("Tangente") # ajoute un titre
#ax1.legend()
ax1.set_xlabel("abscisses")
ax1.set_ylabel("ordonnées")

plt.show()
```

# Crée son module
il faut le faire dans un fichier py

```python
%%writefile tools_matTP2.py
import numpy as np
import numpy.random as rd

def Matlap(n):
    
    """Envoie la mtrice du laplacien 2D à n lignes et n colonnes"""
    A = 2*np.eye(n)
    v = np.ones(n-1)
    A -= np.diag(v,1) + np.diag(v, -1)
    return A

def MatSymRand(n):
    "renvoei la matrice symetrice n*n aléatoire"
    A = rd.random_sample((n,n))
    return A+ A.transpose()

if(__name__=="__main__"):
    print("Matrice symétrique aléatoire avec MatSymRand")
    print(" ", MatSymRand(4))
```

```python
import tools_matTP2 as mt
print(mt.Matlap(10))
print(mt.MatSymRand(10))
```
