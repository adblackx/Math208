```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Introduction

Le module `matplotlib` permet de tracer des fonctions. Par exemple, pour tracer $x \mapsto \sqrt{x}$, le code minimal est le suivant :

```python
# D√©finition des valeurs pour l'abscisse
x = np.linspace(0,10,5)

# Cr√©ation d'une figure contenant un graphe
fig, ax = plt.subplots() 
# Ajout de la courbe
ax.plot(x,np.sqrt(x))
```

La courbe consiste √† "relier" les points de coordonn√©es $(x[i],y[i])$ ; sur l'exemple ci-dessous, il n'y pas assez de points donc la courbe n'est pas assez arrondie. Si l'on prend plus de points, on reconna√Æt mieux la fonction racine carr√©e :

```python
# D√©finition des valeurs pour l'abscisse
x = np.linspace(0,10,150)

# Cr√©ation d'une figure contenant un graphe
fig, ax = plt.subplots() 
# Ajout de la courbe
ax.plot(x,np.sqrt(x),label="racine carr√©e")# On trace une premi√®re courbe, le label sert pour la l√©gende
ax.plot(x,x,label="identit√©")# On trace une seconde courbe.
ax.legend() # Obligatoire pour afficher les l√©gendes
# Ajout des titres des axes :
ax.set_xlabel("x") # titre de l'axe des  abscisses
ax.set_ylabel("x") # titre de l'axe des ordonn√©es
ax.set_title("mes des courbes") # titre du graphe
```

Il est possible d'afficher plusieurs graphes (c√¥te-√†-c√¥te, ou sur plusieurs lignes). Tester le code suivant :

```python
def f(x):
    return np.sin(x)
    
def test_plot():
    s =0.
    e = 4.*np.pi
    x = np.linspace(s,e, 150)
    y = f(x)
    fig, ax = plt.subplots(2,1,figsize=(10,5))
    ax[0].plot(x, y,label="ma courbe")
    ax[0].set_title(r'$y=\sin(x)$')
    ax[0].set_xlabel("x")
    ax[0].set_ylabel("sinus")
    ax[0].legend()
    
    ax[1].plot(x, np.cos(x),label="ma courbe")
    ax[1].set_title(r'$y=\cos(x)$')
    ax[1].set_xlabel("x")
    ax[1].set_ylabel("cosinus")
    ax[1].legend()
    
    fig.suptitle("Fonctions trigonom√©triques √©l√©mentaires")
    #fig.show() # Inutile dans les notebooks uniquement.

test_plot()
```

### Question
1. Pour $t\in [0;2\pi]$, tracer la courbe param√©tr√©e :
\begin{equation}
\left\{
\begin{aligned}
x(t) &= \sin^3 (t) \\
y(t) &= \cos(t) - \cos^4 (t)
\end{aligned}
\right.

2. Tracer la courbe de Lissajou suivante :

\end{aligned}

```python
# Mettre votre code ici
# D√©finition des valeurs pour l'abscisse
t = np.linspace(0,2*np.pi,1000)

xt = np.sin(t)**3
yt = np.cos(t) - np.cos(t)**4

plt.plot(xt, yt) ## PAS SUR SUR CETTE METHODE
plt.axis("equal")
plt.xlim(-3,3)
plt.show()

xt1 = np.sin(5*t)
yt1 = np.cos(3*t+np.pi)

plt.plot(xt1, yt1) ## PAS SUR SUR CETTE METHODE
plt.axis("equal")
plt.xlim(-3,3)

plt.show()
```

## Param√©trisation polaire
Cette param√©trisation consiste √† rep√©rer un point en fonction d'un angle (le z√©ro correspondant √† l'axe des abscisse) et d'un rayon (distance √† l'origine de l'axe), par exemple l'angle $\theta$ pour un cercle $x = rcos(\theta)$, $y = rsin(\theta)$.
Matplotlib fournit _nativement_ des outils pour tracer des courbes en coordonn√©es polaires.

Pour pouvoir d√©finir une courbe √† l'aide des coordonn√©es polaires, il faut utiliser la projection polaire : `projection='polar'`. En voici deux exemples :

```python
t = np.linspace(0,2.*np.pi,200)

# Exemple 1 : uniquement des polaires
fig, ax = plt.subplots(1,2,figsize=(10,2),subplot_kw=dict(projection='polar'))
ax[0].plot(t,np.ones(t.shape))
ax[1].plot(np.sin(t),np.cos(t))
fig.subplots_adjust(top=0.8,wspace=0.) # Pour ajuster l'espacement (wspace pour horizontal, hspace pour vertical)
fig.suptitle("Uniquement des courbes polaires")

# Exemple 2 : m√©lange
fig2 = plt.figure()
ax0 = fig2.add_subplot(121, projection='polar')
ax1 = fig2.add_subplot(122, aspect='equal')
ax0.plot(t,np.ones(t.shape))
ax1.plot(np.sin(t),np.cos(t))
fig2.subplots_adjust(top=1.1,wspace=0.5) # Pour ajuster l'espacement (wspace pour horizontal, hspace pour vertical)
fig2.suptitle('Une courbe en polaire et une "classique"')
[]
```
