```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,150)

# Cration d'une figure contenant un graphe
fig, ax = plt.subplots(1,1, figsize=(10,5)) 
# Ajout de la courbe
ax.plot(x,np.exp(-x/2),label="racine carre")# On trace une premire courbe, le label sert pour la lgende
ax.plot(x,np.cos(2*x*np.pi),label="racine carre")# On trace une premire courbe, le label sert pour la lgende
ax.legend() # Obligatoire pour afficher les lgendes
# Ajout des titres des axes :
ax.set_xlabel("x") # titre de l'axe des  abscisses
ax.set_ylabel("x") # titre de l'axe des ordonnes
ax.set_title("mes des courbes") # titre du graphe
```
